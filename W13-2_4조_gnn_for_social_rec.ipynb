**GNN? 그래프 신경망 (Graph Neural Network)**
-	***GNN은 그래프 구조의 데이터를 입력으로 사용하는 인공신경망, 가장 많이 쓰이는 인공 신경망들인 FCN이나 RNN은 벡터나 행렬 형태의 입력을 사용하는데, 그에 비해 GNN은 입력이 그래프 구조입니다.*** 그래프 형태의 데이터에서 학습하는 머신러닝 모델 중 하나로, 각 노드와 엣지 간의 상호작용을 모델링합니다.***그래프는 노드들의 집합과 이들을 연결하는 엣지로 구성되어 있습니다. GNN은 이러한 그래프 구조를 효과적으로 학습하여 노드의 특징을 추출하고 예측하는 데 사용됩니다.***
-	GNN의 주요 아이디어는 각 노드가 자신의 이웃들과의 상호작용을 통해 정보를 업데이트한다는 것입니다. 각 노드는 이웃 노드의 정보를 종합하여 자신의 표현을 업데이트하고, 이 과정을 여러 번 반복함으로써 전역적인 그래프 구조를 반영하는 표현을 얻을 수 있습니다.
- ***GNN은 주로 연결관계와 이웃들의 상태를 이용하여 각 점의 상태를 업데이트(학습)하고 마지막 상태를 통해 예측 여부를 수행한다.***

"Graph Neural Networks for Social Recommendation"은 미국의 연구자들이 2019년에 발표한 논문입니다. 이 논문에서는 소셜 네트워크 정보를 활용하여 추천 시스템을 개선하기 위한 그래프 신경망(Graph Neural Network, GNN)을 소개합니다.

***여기서 주요 아이디어는 소셜 네트워크에서의 사용자 간의 관계를 그래프로 표현하고, 이 그래프를 활용하여 추천 시스템을 개선하는 것입니다. 일반적인 추천 시스템은 사용자의 과거 행동이나 아이템의 특성을 기반으로 추천을 수행합니다. 그러나 소셜 네트워크 정보는 사용자 간의 사회적 연결성을 제공하므로 이를 활용하여 더 정확하고 다양한 추천을 할 수 있습니다.***


1. 서론:

논문의 주요 목적은 소셜 네트워크 정보를 활용하여 추천 시스템을 개선하는 것입니다.
본 논문에서는 소셜 추천을 위한 새로운 그래프 신경망 프레임워크, GraphRec를 제시합니다.
소셜 네트워크에서의 사용자 간의 관계를 그래프로 모델링하고, 이를 기반으로 추천 모델을 강화하는 방식을 제안합니다.

2. 배경:

기존 추천 시스템은 주로 사용자의 과거 행동이나 아이템 특성을 기반으로 동작하며, 이로부터 한계점이 발생합니다.
소셜 네트워크 정보는 사용자 간의 사회적 연결성을 나타내므로, 이를 활용하여 추천의 다양성과 정확도를 향상시킬 수 있습니다.

문제점(해결해야하는 과제)
- 사용자-아이템 그래프와 사용자-사용자 소셜 그래프를 어떻게 결합할 것인가
- 사용자와 아이템 간의 상호작용을 어떻게 포착할 것인가
- 이질적인 강점을 지닌 사회적 관계를 어떻게 구별할 것인가

3. 방법:

***그래프 신경망(GNN)을 사용하여 소셜 그래프의 정보를 효과적으로 학습합니다.
GNN을 통해 사용자 간의 관계를 모델링하고, 이를 기반으로 아이템 추천을 수행합니다.***
***사용자와 아이템 간의 상호 작용을 그래프 형태로 나타내고, GNN을 통해 이를 반영하여 추천 결과를 개선합니다.***


4. 결과:

***실험 결과는 소셜 네트워크 정보를 활용한 GNN 기반 추천 시스템이 기존 방법보다 더 나은 성능을 보인다는 것을 보여줍니다.***
***특히, 사용자 간의 관계를 고려함으로써 추천의 다양성과 정확성이 향상됨을 확인할 수 있습니다.***

5. 결론:

소셜 네트워크 정보를 활용한 GNN은 추천 시스템에서 유용하게 적용될 수 있는 유망한 방법론임을 제시합니다.
***사용자의 사회적 연결성을 고려하는 것이 추천 시스템의 성능 향상에 기여할 수 있으며, 앞으로의 연구 방향을 제시합니다.***

이 코드는 RMSprop 옵티마이저를 사용하여 모델을 훈련하며, 평가 지표로는 RMSE(Root Mean Square Error)와 MAE(Mean Absolute Error)를 활용합니다. 또한, 조기 종료(Early Stopping)가 구현되어 성능 향상이 미미할 때 훈련을 조기에 종료합니다.

RMSprop 옵티마이저 : 경사 하강법 기반의 최적화 알고리즘으로 지수 이동 평균(Exponential Moving Average) 사용하여 최근 그래디언트에 더 많은 가중치를 주고, 과거는 적은 가중치를 부여.

```python
class GraphRec(nn.Module):
    # GraphRec 모델 클래스의 초기화 부분입니다.
    # 클래스 초기화하는 이유는 모델의 파라미터들을 초기화하고 모델을 훈련시키기 위함 - init (기본값 설정)
    # 각 인코더를 초기화해준다.
    def __init__(self, enc_u, enc_v_history, r2e): #enc_u 사용자 ; enc_v_history 아이템의상호작용 ; r2e 평가를 임베딩으로 변환하는 레이블 인코더
        super(GraphRec, self).__init__()
        self.enc_u = enc_u
        self.enc_v_history = enc_v_history
        self.embed_dim = enc_u.embed_dim

        self.w_ur1 = nn.Linear(self.embed_dim, self.embed_dim) #사용자 노드의 임베딩을 처리하는 두개의 선형레이어
        self.w_ur2 = nn.Linear(self.embed_dim, self.embed_dim)
        self.w_vr1 = nn.Linear(self.embed_dim, self.embed_dim) #아이템 노드의 임베딩을 처리하는 두개의 선형레이어
        self.w_vr2 = nn.Linear(self.embed_dim, self.embed_dim)
        self.w_uv1 = nn.Linear(self.embed_dim * 2, self.embed_dim) #사용자와 아이템 임베딩을 결합하는데 사용되는 선형레이어
        self.w_uv2 = nn.Linear(self.embed_dim, 16) #이전 레이어의 출력을 16차원으로 변환
        self.w_uv3 = nn.Linear(16, 1)
        self.r2e = r2e #아이템 레이블을 임베딩으로 변환하는데 사용되는 외부의 임베딩 레이어
        self.bn1 = nn.BatchNorm1d(self.embed_dim, momentum=0.5)
        self.bn2 = nn.BatchNorm1d(self.embed_dim, momentum=0.5)
        self.bn3 = nn.BatchNorm1d(self.embed_dim, momentum=0.5)
        self.bn4 = nn.BatchNorm1d(16, momentum=0.5) #batch normalization레이어로, 각 레이어의 출력에 대해서 배치 정규화 수행
        self.criterion = nn.MSELoss()

    # 그래프 모델의 순전파를 정의하는 부분입니다. 각 레이어의 연산이 진행됩니다.
    # (순전파 - 뉴럴 네트워크 모델의 입력층부터 출력층까지 순서대로 변수들을 계산하고 저장하는 것을 의미합니다.)
    # 입력으로 받은 node_u와 node_v에 대해 각각 사용자임베딩 인코더와 아이템history인코더를 통과시켜 임베딩을 얻은 후,
    # 다양한 레이어 및 활성화 함수를 통과시켜 최종 예측값을 생성함

    def forward(self, nodes_u, nodes_v):
        embeds_u = self.enc_u(nodes_u)
        embeds_v = self.enc_v_history(nodes_v)

        x_u = F.relu(self.bn1(self.w_ur1(embeds_u)))
        x_u = F.dropout(x_u, training=self.training)
        x_u = self.w_ur2(x_u)
        x_v = F.relu(self.bn2(self.w_vr1(embeds_v)))
        x_v = F.dropout(x_v, training=self.training)
        x_v = self.w_vr2(x_v)

        x_uv = torch.cat((x_u, x_v), 1)
        x = F.relu(self.bn3(self.w_uv1(x_uv)))
        x = F.dropout(x, training=self.training)
        x = F.relu(self.bn4(self.w_uv2(x)))
        x = F.dropout(x, training=self.training)
        scores = self.w_uv3(x)
        return scores.squeeze()

     # 손실 함수를 정의하는 부분입니다. 예측값과 실제값 간의 평균 제곱 오차를 계산합니다.
    def loss(self, nodes_u, nodes_v, labels_list):
        scores = self.forward(nodes_u, nodes_v)
        return self.criterion(scores, labels_list)
```

```python
# 모델을 훈련하는 함수입니다. 훈련 데이터로 모델을 업데이트하고 손실을 출력합니다.
def train(model, device, train_loader, optimizer, epoch, best_rmse, best_mae):
    model.train()
    running_loss = 0.0 #각 배치의 러닝 손실을 추적하기 위해서 변수를 초기화합니다.
    for i, data in enumerate(train_loader, 0): #train데이터 loader에서 미니배치를 반복적으로 가져옵니다.
        batch_nodes_u, batch_nodes_v, labels_list = data
        optimizer.zero_grad() #기울기를 초기화합니다.
        loss = model.loss(batch_nodes_u.to(device), batch_nodes_v.to(device), labels_list.to(device)) #모델을 통해 예측값을 얻고 손실을 계산합니다.
        loss.backward(retain_graph=True) #역전파를 통해 기울기를 계산합니다.
        optimizer.step() #최적화를 수행하여 모델 파라미터를 업데이트합니다.
        running_loss += loss.item()
        if i % 100 == 0:
            print('[%d, %5d] loss: %.3f, The best rmse/mae: %.6f / %.6f' % (
                epoch, i, running_loss / 100, best_rmse, best_mae))
            running_loss = 0.0
    return 0
```

```python
# 모델을 평가하는 함수입니다. 테스트 데이터에 대한 예측값과 실제값 간의 RMSE 및 MAE를 계산합니다.
def test(model, device, test_loader):
    model.eval() #모델을 평가모드로 설정하고
    tmp_pred = []
    target = []
    with torch.no_grad(): #테스트 데이터로더에서 미니배치를 반복적으로 가져옵니다.
        for test_u, test_v, tmp_target in test_loader:
            test_u, test_v, tmp_target = test_u.to(device), test_v.to(device), tmp_target.to(device)
            val_output = model.forward(test_u, test_v) #모델을 통해서 예측값을 얻습니다.
            tmp_pred.append(list(val_output.data.cpu().numpy()))
            target.append(list(tmp_target.data.cpu().numpy()))
    tmp_pred = np.array(sum(tmp_pred, []))
    target = np.array(sum(target, []))
    expected_rmse = sqrt(mean_squared_error(tmp_pred, target))
    mae = mean_absolute_error(tmp_pred, target)
    return expected_rmse, mae
```

UV_Aggregator

```python
# 아이템과 사용자의 임베딩을 집계하는 클래스
# forward 매서드는 현재 노드와 그 history를 기반으로 새로운 임베딩을 생성
# 임베딩은 주어진 history와 이에 해당하는 레이블을 사용하여 생성
# 이웃노드의 임베딩을 통합하고 attention메커니즘을 사용하여 가중치를 할당함
# 최종적으로 생성된 임베딩 행렬을 반환

import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.nn.functional as F
import numpy as np
import random
#from Attention import Attention


class UV_Aggregator(nn.Module):
    """
    item and user aggregator: for aggregating embeddings of neighbors (item/user aggreagator).
    """

    def __init__(self, v2e, r2e, u2e, embed_dim, cuda="cpu", uv=True):
        super(UV_Aggregator, self).__init__()
        self.uv = uv
        self.v2e = v2e  # 아이템의 임베딩을 위한 nn.Embedding 모듈
        self.r2e = r2e  # 상호 작용 레이블의 임베딩을 위한 nn.Embedding 모듈
        self.u2e = u2e  # 사용자(또는 아이템)의 임베딩을 위한 nn.Embedding 모듈
        self.device = cuda  # 장치 (CPU 또는 GPU)
        self.embed_dim = embed_dim  # 임베딩 차원
        self.w_r1 = nn.Linear(self.embed_dim * 2, self.embed_dim)  # 선형 레이어 1
        self.w_r2 = nn.Linear(self.embed_dim, self.embed_dim)  # 선형 레이어 2
        self.att = Attention(self.embed_dim) # Attention 클래스를 사용

    def forward(self, nodes, history_uv, history_r):
      # param nodes : 현재 노드
      # param history_uv: 현재 노드와 상호 작용한 이력 (이웃 노드)
      # param history_r: 각 상호 작용에 대한 레이블
      # return: 새로운 임베딩 행렬

        embed_matrix = torch.empty(len(history_uv), self.embed_dim, dtype=torch.float).to(self.device)

        for i in range(len(history_uv)):
            history = history_uv[i]
            num_histroy_item = len(history)
            tmp_label = history_r[i]

            if self.uv == True:
                # user component
                e_uv = self.v2e.weight[history]
                uv_rep = self.u2e.weight[nodes[i]]
            else:
                # item component
                e_uv = self.u2e.weight[history]
                uv_rep = self.v2e.weight[nodes[i]]

            e_r = self.r2e.weight[tmp_label]
            x = torch.cat((e_uv, e_r), 1)
            x = F.relu(self.w_r1(x))
            o_history = F.relu(self.w_r2(x))

            att_w = self.att(o_history, uv_rep, num_histroy_item)
            att_history = torch.mm(o_history.t(), att_w)
            att_history = att_history.t()

            embed_matrix[i] = att_history
        to_feats = embed_matrix
        return to_feats
```

Social_Aggregator


```python
 # Social Aggregator는 사용자의 소셜 네트워크에서 얻은 이웃들의 임베딩을 집계하는 역할

import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np
import random
#from Attention import Attention


class Social_Aggregator(nn.Module):
    """
    Social Aggregator: for aggregating embeddings of social neighbors.
    """

    ##Social Aggregator의 초기 설정을 수행
    def __init__(self, features, u2e, embed_dim, cuda="cpu"):
        super(Social_Aggregator, self).__init__()

        self.features = features
        self.device = cuda
        self.u2e = u2e
        self.embed_dim = embed_dim
        self.att = Attention(self.embed_dim)

    ##forward 메서드는 Social Aggregator의 순전파 연산을 정의
    def forward(self, nodes, to_neighs):
        ##embed_matrix는 현재 중심 사용자 노드에 대한 이웃들의 임베딩을 저장하기 위한 행렬
        embed_matrix = torch.empty(len(nodes), self.embed_dim, dtype=torch.float).to(self.device)
        for i in range(len(nodes)):
            tmp_adj = to_neighs[i]
            num_neighs = len(tmp_adj)
            #
            e_u = self.u2e.weight[list(tmp_adj)] # fast: user embedding
            #slow: item-space user latent factor (item aggregation)
            #feature_neigbhors = self.features(torch.LongTensor(list(tmp_adj)).to(self.device))
            #e_u = torch.t(feature_neigbhors)

            u_rep = self.u2e.weight[nodes[i]]

            att_w = self.att(e_u, u_rep, num_neighs)
            att_history = torch.mm(e_u.t(), att_w).t()
            embed_matrix[i] = att_history
        to_feats = embed_matrix

        return to_feats
```

UV_Encoder

```python
import torch
import torch.nn as nn
from torch.nn import init
import torch.nn.functional as F


class UV_Encoder(nn.Module):

    def __init__(self, features, embed_dim, history_uv_lists, history_r_lists, aggregator, cuda="cpu", uv=True):
        super(UV_Encoder, self).__init__()

        self.features = features
        self.uv = uv
        self.history_uv_lists = history_uv_lists
        self.history_r_lists = history_r_lists
        self.aggregator = aggregator
        self.embed_dim = embed_dim
        self.device = cuda
        self.linear1 = nn.Linear(2 * self.embed_dim, self.embed_dim)  #

    def forward(self, nodes):
        tmp_history_uv = []
        tmp_history_r = []
        for node in nodes:
            tmp_history_uv.append(self.history_uv_lists[int(node)])
            tmp_history_r.append(self.history_r_lists[int(node)])

        neigh_feats = self.aggregator.forward(nodes, tmp_history_uv, tmp_history_r)  # user-item network

        self_feats = self.features.weight[nodes]
        # self-connection could be considered.
        combined = torch.cat([self_feats, neigh_feats], dim=1)
        combined = F.relu(self.linear1(combined))

        return combined
```

Social_Encoder

```python
import torch
import torch.nn as nn
from torch.nn import init
import torch.nn.functional as F


class Social_Encoder(nn.Module):

    def __init__(self, features, embed_dim, social_adj_lists, aggregator, base_model=None, cuda="cpu"):
        super(Social_Encoder, self).__init__()

        self.features = features
        self.social_adj_lists = social_adj_lists
        self.aggregator = aggregator
        if base_model != None:
            self.base_model = base_model
        self.embed_dim = embed_dim
        self.device = cuda
        self.linear1 = nn.Linear(2 * self.embed_dim, self.embed_dim)  #

    def forward(self, nodes):

        to_neighs = []
        for node in nodes:
            to_neighs.append(self.social_adj_lists[int(node)])
        neigh_feats = self.aggregator.forward(nodes, to_neighs)  # user-user network

        self_feats = self.features(torch.LongTensor(nodes.cpu().numpy())).to(self.device)
        self_feats = self_feats.t()

        # self-connection could be considered.
        combined = torch.cat([self_feats, neigh_feats], dim=1)
        combined = F.relu(self.linear1(combined))

        return combined
```

Attention

```python
import torch
import torch.nn as nn
from torch.nn import init
import numpy as np
import random
import torch.nn.functional as F


class Attention(nn.Module):
    def __init__(self, embedding_dims):
        super(Attention, self).__init__()
        self.embed_dim = embedding_dims
        self.bilinear = nn.Bilinear(self.embed_dim, self.embed_dim, 1)
        self.att1 = nn.Linear(self.embed_dim * 2, self.embed_dim)
        self.att2 = nn.Linear(self.embed_dim, self.embed_dim)
        self.att3 = nn.Linear(self.embed_dim, 1)
        self.softmax = nn.Softmax(0)

    def forward(self, node1, u_rep, num_neighs):
        uv_reps = u_rep.repeat(num_neighs, 1)
        x = torch.cat((node1, uv_reps), 1)
        x = F.relu(self.att1(x))
        x = F.dropout(x, training=self.training)
        x = F.relu(self.att2(x))
        x = F.dropout(x, training=self.training)
        x = self.att3(x)
        att = F.softmax(x, dim=0)
        return att
```
